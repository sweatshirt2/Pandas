```python
import pandas as pd
```

```python
df1 = pd.read_csv(r'resources\LOTR.csv')
df1
```

```python
df2 = pd.read_csv(r'resources\LOTR 2.csv')
df2
```

#### Merge
- kwargs
  - on
    - which columns to merge the dfs on
  - how
    - type of merge
      - inner (natural)
      - outer
      - left
      - right
      - cross

```python
df1.merge(df2)
```

##### <i>Note</i>
- if the on kwarg is specified it the table includes all the other columns even if there are some with similar values

```python
df1.merge(df2, on='FellowshipID')
```

```python
df1.merge(df2, how='left')
```

```python
df1.merge(df2, how='right')
```

```python
df1.merge(df2, how='outer')
```

```python
df1.merge(df2, how='cross')
```

##### Join
- requires suffixes to be specified

```python
df1.join(df2, lsuffix='_1', rsuffix='_2')
```

##### <i>Note</i>
- with no on specified it tries to fill gaps with the specified suffixes

```python
df1.join(df2, on='FellowshipID', rsuffix='_2')
```

##### <i>Note</i>
- specifying 'on' of the suffixes is enough
- default value of 'how' seems to be 'left'

```python
df1.set_index('FellowshipID').join(df2.set_index('FellowshipID'), rsuffix='_2', how='outer')
```

```python
pd.concat([df1, df2])
```

##### <i>Note</i>
- the default value for 'join' which is like 'how' for merge and join seems to be 'outer'

```python
pd.concat([df1.set_index('FellowshipID'), df2.set_index('FellowshipID')], axis=1)
```
