```python
import pandas as pd
```

```python
df = pd.read_csv(r'resources/Ice Cream Ratings.csv', index_col='Date')
df
```

##### The plot method
- by default it creates a line graph | type='line'
- In the plotted diagram the x-axis hows the index column and the y-axis shows the rating range
- Use the kind named argument to change the plot kind

```python
df.plot()
```

```python
df.plot(kind='line', title='Ice Cream Ratings')
```

```python
df.plot(subplots=True, xlabel='Date', ylabel='Rating')
```

```python
df['Flavor Rating'].plot(kind='bar', stacked=True)
```

```python
df.plot(stacked=True)
```

```python
df.plot.bar()
```

##### <i>Note</i>
- make the bar plot using barh

```python
df.plot.barh()
```

```python
df.plot.scatter(x = 'Flavor Rating', y = 'Overall Rating', s=123, c='brown')
```

##### <i>Note</i> 
- unlike the other plots, scatter plot needs x and y to be defined
- named arguments
  - s can be used to specify the size of the scatter points
  - c can be used to specify the color of the scatter points

```python
df.plot.hist(bins=50)
```

```python
df.boxplot()
```

```python
df.plot.area()
```

##### Set the size of the figure using figsize named argument

```python
df.plot.pie(y='Overall Rating', figsize=(10, 10))
```

##### <i>Note</i>
- The pie plot needs to know which column is about to be displayed
