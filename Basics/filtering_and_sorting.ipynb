#### Filtering with filter()
- kwargs
  -  items
     -  selected list of rows or columns
  - like
    -  works like items but values do not have to much the exact input
  - axis
    - default value 1
    - choose whether to select using indexes or titles/names with 0 and 1

```python
df.filter(items=['Capital', 'Growth Rate'])
```

```python
df.filter(items=['Yemen', 'Zambia', 'Albania'], axis=0)
```

```python
df.filter(like='C')
```

```python
df.filter(like='Al', axis=0)
```
